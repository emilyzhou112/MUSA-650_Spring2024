### Example: a more advanced outlier detection

The *anomaly score* of each sample is called the **Local Outlier Factor**. It measures the local deviation of the density of a given sample with respect to its neighbors. It is local in that the anomaly score depends on how isolated the object is with respect to the surrounding neighborhood. More precisely, locality is given by k-nearest neighbors, whose distance is used to estimate the local density. *By comparing the local density of a sample to the local densities of its neighbors, one can identify samples that have a substantially lower density than their neighbors. These are considered outliers.*

```python
# Use the Local Outlier Factor (LOF) algorithm from the scikit-learn library
# for unsupervised outliers detection in a univariate time series.
from sklearn.neighbors import LocalOutlierFactor

# fit the model for outlier detection (default)
X = np.array(dfTmp.tsRel).reshape(dfTmp.shape[0],1) # convert tsRel into an array and convert this array tfrom 1D to 2D
X.shape # (19,1)
# Create an instance of the LocalOutlierFactor class
# where the number of neighbors to consider for each sample is 5
# contamination is an estimate or assumption about the percentage of anomalies in the data.
clf = LocalOutlierFactor(n_neighbors=5, contamination=0.1)
clf.fit_predict(X) # fits the model to the data (X) and assigns labels (+1 for inliers and -1 for outliers) to each data point.

# assign the negative outlier factor scores to each data point.
# The negative outlier factor is a measure of the anomaly score,
# and lower values typically indicate more anomalous points.
dfTmp['outScore'] = clf.negative_outlier_factor_.tolist()
dfTmp
```

```python
dfTmp.plot.scatter(x='tsRel', y='outScore', c='DarkBlue') # plot outscore
```

The following PCAs are conducted

- Using normalized data

`mdlPCA.explained_variance_ratio_`: This prints the explained variance ratio for each of the selected principal components. The explained variance ratio represents the proportion of the dataset's variance that is explained by each principal component.
`np.sum(mdlPCA.explained_variance_ratio_)`: This prints the cumulative sum of the explained variance ratios. It indicates the total amount of variance explained by the selected principal components.

```python
# Performs PCA on df2
mdlPCA = PCA(n_components=5) # the PCA will retain only the top 5 principal components
XPCA = mdlPCA.fit_transform(df2)

print(mdlPCA.explained_variance_ratio_)
print(np.sum(mdlPCA.explained_variance_ratio_))

# This line plots the explained variance ratio for each principal component.
# The x-axis represents the index of the principal component (from 0 to the number of components - 1), and the y-axis represents the explained variance ratio.
plt.plot(np.arange(0,mdlPCA.explained_variance_ratio_.shape[0]), mdlPCA.explained_variance_ratio_, '-ro')
```

### Normalize data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

#scaler = StandardScaler()
scaler = MinMaxScaler() # define a scaler
# The MinMaxScaler is chosen for this particular case.
# It scales features to a specified range, typically between 0 and 1.

scaler.fit(X) # scale X now using the imported library instead of defining own function
Xnorm = scaler.transform(X)
Xnorm.max(axis=0)
```

```python
from sklearn.svm import SVR
```

#### Cross validation with leave one out

```python
from sklearn.model_selection import LeaveOneOut

# perform the leave one group out cross validation
loo = LeaveOneOut()
for train_index, test_index in loo.split(Xnorm):
    print('TRAIN: ' + str(train_index) + '  TEST: ' +  str(test_index))
```

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.svm import LinearSVR

loo = LeaveOneOut() # define the CV method

predAll = np.zeros([y.shape[0],1])


# for loop below for LOGO cross validation
i=0

for train_index, test_index in loo.split(Xnorm):

    # define training and testing
    X_train, X_test = Xnorm[train_index], Xnorm[test_index]
    y_train, y_test = y[train_index], y[test_index]

    regr = LinearSVR(random_state=0, tol=1e-5) # linear regression model
    regr.fit(X_train, y_train)  #  Train the model
    ypred = regr.predict(X_test)  #  Apply the model

    predAll[i] = ypred

    print('Y : ' + str(y_test) + ' , pred as: ' + str(ypred)) # actual versus predicted value
    i = i + 1
```

```python
np.corrcoef(y.T, predAll.T)
```
